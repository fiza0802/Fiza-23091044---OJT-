```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(44)
P= np.random.randint(-10, 11, size=6)

def polynomial(x):
    return np.polyval(P, x)
x_val = np.linspace(-5, 5, 400)
y_val = polynomial(x_val)
plt.figure(figsize=(10, 6))
plt.plot(x_val, y_val, label='Polynomial', color='b')

# Find maxima and minima
maxima_indices = np.r_[True, y_val[1:] > y_val[:-1]] & np.r_[y_val[:-1] > y_val[1:], True]
minima_indices = np.r_[True, y_val[1:] < y_val[:-1]] & np.r_[y_val[:-1] < y_val[1:], True]

maxima_x = x_val[maxima_indices]
maxima_y = y_val[maxima_indices]
minima_x = x_val[minima_indices]
minima_y = y_val[minima_indices]

# Annotate maxima and minima
for i in range(len(maxima_x)):
    plt.annotate('Max', xy=(maxima_x[i], maxima_y[i]), xytext=(maxima_x[i] + 0.5, maxima_y[i] + 10),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

for i in range(len(minima_x)):
    plt.annotate('Min', xy=(minima_x[i], minima_y[i]), xytext=(minima_x[i] - 0.5, minima_y[i] - 10),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.title('Plot of 5th Degree Polynomial with maxima and minima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
